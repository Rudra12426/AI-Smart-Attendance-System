```python
import cv2
import os
import pandas as pd
from datetime import datetime
from deepface import DeepFace
```

```python
ATTENDANCE_FILE = r"C:\Users\Victus\OneDrive\Desktop\ML Project\AI Attendance System\ATENDANCE_FILE"

if not os.path.exists("attendance"):
    os.makedirs("attendance")

if not os.path.exists(ATTENDANCE_FILE):
    df = pd.DataFrame(columns=["Name", "Date", "Time"])
    df.to_csv(ATTENDANCE_FILE, index=False)
```

```python
def mark_attendance(name):
    df = pd.read_csv(ATTENDANCE_FILE)
    today = datetime.now().strftime("%Y-%m-%d")

    if not ((df["Name"] == name) & (df["Date"] == today)).any():
        time_now = datetime.now().strftime("%H:%M:%S")
        df.loc[len(df)] = [name, today, time_now]
        df.to_csv(ATTENDANCE_FILE, index=False)
        print(f"âœ… Attendance marked for {name}")

```

```python
cap = cv2.VideoCapture(0)
print("ðŸŽ¥ Camera started")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    try:
        results = DeepFace.find(
            img_path=frame,
            db_path="dataset",
            enforce_detection=False,
            silent=True
        )

        if len(results) > 0 and len(results[0]) > 0:
            identity = results[0].iloc[0]["identity"]
            name = identity.split(os.sep)[-2]

            mark_attendance(name)

            cv2.putText(frame, name, (50, 50),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        1, (0, 255, 0), 2)

    except Exception as e:
        pass
```

```python
cv2.imshow("AI Smart Attendance System",frame)
 
 if cv2.waitKey(1) &0xFF ==ord('Esc'):
  break

cap.release()
cv2.destroAllWindows()
print("System Closed")
```
